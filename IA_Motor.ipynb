```python
def mostrar_valores_faltantes(df):
    for col in df.columns.tolist():
        print('{} column missing values: {}'.format(col, df[col].isnull().sum()),'. (',(df[col].isnull().sum()/df.shape[0])*100,'%)')
    print('\n')
mostrar_valores_faltantes(ds)
```

```python
def Eliminadatosfaltantes (dataset,columna):
  dataset = dataset[dataset[columna].notna()]
  return dataset

ds =Eliminadatosfaltantes(ds,"review_content")
```

```python
getcolumns = ds.drop(columns=['rotten_tomatoes_link','critic_name','top_critic','publisher_name','review_score','review_date'])
getcolumns.describe()
```

```python
#80% de los registros
df_80 = getcolumns.sample(frac=0.90, random_state=200)

# obtener los faltantes
df_20 = getcolumns[~getcolumns.index.isin(df_80.index)]

df_80.describe()
df_20.describe()
```

```python
rottenNarray = df_80["review_content"].loc[df_80["review_type"] == 'Rotten'].to_numpy()
rotten =  rottenNarray.astype(str).tolist()
freshNarray = df_80["review_content"].loc[df_80["review_type"] == 'Fresh'].to_numpy()
fresh = freshNarray.astype(str).tolist()
```

```python
fresh
```

```python
total_sentences = len(rotten) + len(fresh)
p_rotten = len(rotten) / total_sentences
p_fresh = len(fresh) / total_sentences
print(p_rotten, p_fresh)   
```

```python
def crear_tabla_frecuencias(corpus, corpus2):
    frecuencias = {}
    for oracion in corpus:
        tokens = oracion.split(" ")
        for token in tokens:
            if token not in frecuencias.keys():
                frecuencias[token] = 1
            else:
                frecuencias[token] += 1
    for oracion in corpus2: #Agregar las palabas no encontradas en el otro vector para suavizado de laplace
        tokens = oracion.split(" ")
        for token in tokens:
            if token not in frecuencias.keys():
                frecuencias[token] = 0
    return frecuencias
frecuencia_rotten = crear_tabla_frecuencias(rotten, fresh)
frecuencia_fresh = crear_tabla_frecuencias(fresh, rotten)

print(len(frecuencia_rotten), len(frecuencia_fresh))
```

```python
def sumar_constante(tabla_frec, constante):
    for k in tabla_frec:
        tabla_frec[k] +=1
    return tabla_frec
sumar_constante(frecuencia_rotten,1)
sumar_constante(frecuencia_fresh,1)
```

```python
def contar_palabras(corpus):
    frecuencia = 0
    for oracion in corpus:
        frecuencia += len(oracion.split(" "))
    return frecuencia
total_rotten = contar_palabras(rotten)
total_fresh = contar_palabras(fresh)
print(total_rotten,total_fresh)
```

```python
def transformar_frecuencia_probabilidad(frecuencias, total, constante):
    cpt_equivalente = {}
    for  k,v in frecuencias.items():
        probabilidad = v / (total + (len(frecuencias)*constante))
        cpt_equivalente[k] = probabilidad
    return cpt_equivalente

cpt_rotten = transformar_frecuencia_probabilidad(frecuencia_rotten, total_rotten, 1)
cpt_fresh = transformar_frecuencia_probabilidad(frecuencia_fresh, total_fresh, 1)
```

```python
def generacion_querys_rotten(mensaje):
    palabras = mensaje.split(" ")
    rottenq = "(p_rotten"
    freshq ="(p_fresh"
    counter = len(palabras)
    for palabra in palabras:
        if(cpt_rotten.get(palabra) != None and cpt_fresh.get(palabra) != None ):  
            rottenq += "*"
            freshq +="*"
            rottenq += "cpt_rotten[" + "'" + palabra + "']"
            freshq += "cpt_fresh[" + "'" + palabra + "']"    
        counter -=1
        if(counter == 0):
            rottenq += ")"
            freshq +=")"
    query = rottenq + "/(" + rottenq +"+" + freshq + ")"
    return eval(query)

def generacion_querys_fresh(mensaje):
    palabras = mensaje.split(" ")
    rottenq = "(p_rotten"
    freshq ="(p_fresh"
    counter = len(palabras)
    for palabra in palabras:
        if(cpt_rotten.get(palabra) != None and cpt_fresh.get(palabra) != None):  
            rottenq += "*"
            freshq +="*"
            rottenq += "cpt_rotten[" + "'" + palabra + "']"
            freshq += "cpt_fresh[" + "'" + palabra + "']"
        counter -=1
        if(counter == 0):
            rottenq += ")"
            freshq +=")"

    query = freshq + "/(" + freshq +"+" + rottenq + ")"
    return eval(query)
#tenthrate
fresh_final_prob = (generacion_querys_fresh('bad movie'))
rotten_final_prob = (generacion_querys_rotten('bad movie'))
```

```python
print(fresh_final_prob, rotten_final_prob)
```

```python
if(fresh_final_prob >= rotten_final_prob):
    print("is fresh")
else:
    print("is rotten")
```

```python
numpyEvalRes = df_20['review_type'].to_numpy()
numpyEvalEntries = df_20['review_content'].to_numpy()
evalRes = numpyEvalRes.astype(str).tolist()
evalEntries = numpyEvalEntries.astype(str).tolist()
```

```python
matrizConfusion = {"FreshVerdadero":0,"FreshFalso":0,"RottenVerdadero":0,"RottenFalso":0}
print(matrizConfusion)
```

```python
palabrasDesconocidas = 0
for i in range(len(evalEntries)):
    #try:
    y_hat = generacion_querys_fresh(evalEntries[i])
    if(y_hat >=0.5):
        result = "Fresh"
        if(result==evalRes[i]):
            matrizConfusion["FreshVerdadero"] +=1
        else:
            matrizConfusion["FreshFalso"]+=1
    else:
        result = "Rotten"
        if(result==evalRes[i]):
            matrizConfusion["RottenVerdadero"] +=1
        else:
            matrizConfusion["RottenFalso"]+=1
    #except:
        #palabrasDesconocidas +=1
print(palabrasDesconocidas)
```

```python
matrizConfusion
```

```python
accuracy = (matrizConfusion["FreshVerdadero"]+matrizConfusion["RottenVerdadero"])/(matrizConfusion["FreshVerdadero"]+matrizConfusion["RottenVerdadero"]+matrizConfusion["FreshFalso"]+matrizConfusion["RottenFalso"])
freshprecision = (matrizConfusion["FreshVerdadero"])/(matrizConfusion["FreshVerdadero"]+matrizConfusion["FreshFalso"])
rottenprecision = (matrizConfusion["RottenVerdadero"])/(matrizConfusion["RottenVerdadero"]+matrizConfusion["RottenFalso"])
freshrecall = (matrizConfusion["FreshVerdadero"])/(matrizConfusion["FreshVerdadero"]+matrizConfusion["RottenFalso"])
rottenrecall = (matrizConfusion["RottenVerdadero"])/(matrizConfusion["RottenVerdadero"]+matrizConfusion["FreshFalso"])
freshf1 = 2*((freshprecision*freshrecall)/(freshprecision+freshrecall))
rottenf1 = 2*((rottenprecision*rottenrecall)/(rottenprecision+rottenrecall))

print("Accuracy: " + str(accuracy))
print("Fresh: " +"Precision: " + str(freshprecision)+ " Recall: " + str(freshrecall) + " F1: " +str(freshf1))
print("Rotten: " +"Precision: " + str(rottenprecision) + " Recall: " + str(rottenrecall) + " F1: " +str(rottenf1))
```
